```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://finance.yahoo.com/most-active"

response=requests.get(url)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
stock_table=soup.find("tbody")
stock_table
```

```python
for i in stock_table.find_all("tr"):
  print(i)
```

```python
symbol=[]
for i in stock_table.find_all("tr"):
  sym=i.find("td",attrs={"aria-label":"Symbol"})
  symbol.append(sym.text)
```

```python
symbol
```

```python
names=[]
for i in stock_table.find_all("tr"):
  name=i.find("td",attrs={"aria-label":"Name"})
  names.append(name.text)
```

```python
names
```

```python
prices=[]
for i in stock_table.find_all("tr"):
  price=i.find("td",attrs={"aria-label":"Price (Intraday)"})
  prices.append(price.text)
```

```python
prices
```

```python
per_change=[]
for i in stock_table.find_all("tr"):
  p=i.find("td",attrs={"aria-label":"% Change"})
  per_change.append(p.text)
```

```python
per_change
```

```python
data={"Symbol":symbol,"Name":names,"Price (Intra day)":prices,"% Change":per_change}
```

```python
#data
```

```python
#pip install pandas
```

```python
import pandas as pd
df=pd.DataFrame(data)
df
```

```python
df.to_csv("Most_active_stocks.csv",index=False)
```
